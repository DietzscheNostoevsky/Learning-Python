```python
# Incrreasing the efficacy of OCR 

from PIL import Image 
img = Image.open("Files/home/jovyan/work/readonly/Noisy_OCR.PNG")
display(img)
#The image is noisy 

```

```python
# would OCR work on this ? 
import pytesseract
text = pytesseract.image_to_string(img)
print(text)
```

```python
# OCR didnt work correctly 
# resizong image 

import PIL 
basewidth = 600 # in pizels 
# preserve the aspect ration of the image 

wpercent = basewidth/float(img.size[0])
hsize = int(float(img.size[1])*float(wpercent)) # 600:? = w:l -> ? = wpercent * l 
# resizing using anti-aliasing 

img = img.resize((basewidth,hsize), PIL.Image.ANTIALIAS) # takes an optional filter

display(img) 
img.save('resized_noise.png')
text = pytesseract.image_to_string(img)
print(text)
```

```python
# no improvement in OCR
# now convert to greyscale 
img = Image.open("resized_noise.png")
img = img.convert('L')
img.save('greyscale_noise.png')
display(img)
text = pytesseract.image_to_string(img)
print(text)
```
