```python
# OCR didnt work correctly 
# resizong image 

import PIL 
basewidth = 600 # in pizels 
# preserve the aspect ration of the image 

wpercent = basewidth/float(img.size[0])
hsize = int(float(img.size[1])*float(wpercent)) # 600:? = w:l -> ? = wpercent * l 
# resizing using anti-aliasing 

img = img.resize((basewidth,hsize), PIL.Image.ANTIALIAS) # takes an optional filter

display(img) 
img.save('resized_noise.png')
text = pytesseract.image_to_string(img)
print(text)
```

```python
img = Image.open('Files/home/jovyan/work/readonly/Noisy_OCR.PNG').convert('1') # 1 is passed as string to binarize 
img.save('black_white_noise.jpg')
display(img)

# dotted black and white image 
```

```python
# Binarization function 

def binarize(image_to_binarize, threshold): 
    # converting the input image to greyscale for conversion to binary
    
    output_image = image_to_binarize.convert('L') # greyscale conversin
    
    # interate all pixels of an image, check the threshold, change image intensity
    
    for x in range(output_image.width):
        for y in range(output_image.height):
            if output_image.getpixel((x,y)) < threshold:
                output_image.putpixel((x,y), 0) 
            else :
                output_image.putpixel((x,y), 255) # 8 bit image ?
    return output_image

            
    
```

```python
# cecking binarization functionm 

# create a range of threhhold 

for thresh in  range(0,257,64): # would generate 5 differnt threshold values 
    print(f'Trying with threshold value : {thresh} ')
    
    display(binarize(Image.open('Files/home/jovyan/work/readonly/Noisy_OCR.PNG'), thresh)) # better to first cache the image to avoid reading it from disk
    print('OCR text : ' , pytesseract.image_to_string(
        binarize(Image.open('Files/home/jovyan/work/readonly/Noisy_OCR.PNG'), thresh)))
    
```

```python
# binarization worked best for OCR with changing threshold 
# o turns everything white, 255 everything black , 
# higher threholds brighs out shapes, boldens text 

```
